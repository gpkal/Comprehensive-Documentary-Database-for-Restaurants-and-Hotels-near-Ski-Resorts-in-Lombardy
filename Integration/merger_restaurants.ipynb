```python
import csv
import re
import pandas as pd
import json
import numpy as np
```

```python
# load dei datset
df1 = pd.read_csv("thefork_merged.csv")
df2 = pd.read_csv("ristoranti_google.csv")
df4 = pd.read_csv("advisor_merged.csv")
```

```python
df2['google_ratings'] = df2['google_ratings'].str.replace(',', '.')
```

```python
# rinamino le colonne dei dataset in modo che siano corrispondenti
df1 = df1.rename(columns={'Titles' : 'titles', 'Indirizzo': 'locations', 'nearest_reference': 'reference', 'Città': 'city',
                            'the_fork_ratings':'rating', 'the_fork_prices':'price', 'the_fork_comments':'comments' })
df2 = df2.rename(columns = {'città': 'reference', 'google_ratings':'rating','google_reviews':'comments', 'google_price':'price'})
df4 = df4.rename(columns = {'nearest_reference': 'reference', 'trip_rating': 'rating', 'trip_price': 'price'})
```

```python
#concateno i due dataset
# Unione dei due dataset tramite concatenazione
df = pd.concat([df1, df4], axis = 0,  ignore_index=True)
df
```

```python
#unione con il terzo dataset
df_final = pd.concat([df, df2], axis = 0,  ignore_index=True)
df_final
```

```python
# Sostituzione della virgola con uno spazio solo nei record che contengono la virgola
df_final['locations'] = df_final['locations'].str.replace(',', ' ', regex=True)

# Rimuovi gli spazi extra dalle colonne "titles"
df_final['titles'] = df_final['titles'].str.strip()

# Rimuovi gli spazi extra dalle colonne "comune" nel secondo dataset
df_final['locations'] = df_final['locations'].str.strip()
df_final
```

```python
duplicati = df_final.duplicated(subset=['reference', 'titles'], keep=False)
```

```python
df_unici = df_final[~duplicati].copy()  # Crea un nuovo DataFrame con i record unici

duplicati_gruppi = df_final[duplicati].groupby(['reference', 'titles'])  # Raggruppa i duplicati

for gruppo, dati in duplicati_gruppi:
    record_unificato = dati.iloc[0].copy()  # Copia il primo record del gruppo

    # Converti i valori in stringhe e splitta i valori separati da virgola in liste
    record_unificato['data_source'] = [str(x).split(',') for x in dati['data_source']]
    record_unificato['price'] = [str(x).split(',') for x in dati['price']]
    record_unificato['rating'] = [str(x).split(',') for x in dati['rating']]

    df_unici = df_unici.append(record_unificato, ignore_index=True)
df = df_unici  # Il DataFrame finale con i record unificati
```

```python
# Conversione del dataframe in formato dizionario con orient='records'
data_dict = df_unici.replace(np.nan, None).to_dict(orient='records')


# Conversione del dizionario in formato JSON
json_data = json.dumps(data_dict, default=str, skipkeys=True)
```

```python
# Salvataggio del JSON in un file
with open('file.json', 'w') as json_file:
    json_file.write(json_data)
```

```python

# Rimozione degli attributi con valore "null"
def remove_null_values(data):
    if isinstance(data, dict):
        return {k: remove_null_values(v) for k, v in data.items() if v is not None}
    elif isinstance(data, list):
        return [remove_null_values(v) for v in data if v is not None]
    else:
        return data

# Rimozione degli attributi nulli dal JSON
json_modified = json.dumps(remove_null_values(json_data), default=str, skipkeys=True)

# Parsing del JSON modificato in un oggetto Python
data = json.loads(json_modified)

# Stampa dell'oggetto Python risultante
print(data)
```

```python
with open('file_2.json', 'w') as json_file:
    json_file.write(data)
```

```python
# Carica il file JSON
with open('file_2.json', 'r') as f:
    json_data = json.load(f)


# Itera su ciascun record
for record in json_data:
    # Crea una copia del record
    record_cleaned = record.copy()

    # Rimuovi gli attributi con valore null
    for key, value in record.items():
        if value is None:
            del record_cleaned[key]

    # Sovrascrivi il record originale con il record pulito
    record.clear()
    record.update(record_cleaned)

# Salva il file JSON pulito
with open('file_cleaned.json', 'w') as f:
    json.dump(json_data, f, indent=4)
```
