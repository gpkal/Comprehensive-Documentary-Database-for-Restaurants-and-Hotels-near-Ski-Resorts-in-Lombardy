```python
import csv
import re
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv("api_merged.csv")
df
```

```python
df3 = pd.read_csv('comprensori_modificato.csv')
df3
```

```python
df = df.rename(columns = {'nearest_reference':'reference'})
```

```python
# Conversione degli attributi in minuscolo
df3['comune'] = df3['comune'].str.lower()
df['reference'] = df['reference'].str.lower()

# Creazione di un dizionario per la mappatura tra 'comune' e 'denominazione' in df3
mapping = df3.set_index('comune')['localita'].to_dict()

# Aggiunta dell'attributo 'denominazione' a merged_df utilizzando il metodo map()
df['localita'] = df['reference'].map(mapping)

df
```

```python
df.to_csv('booking_final.csv', index=False)
```

```python
# Conversione del dataframe in formato dizionario con orient='records'
data_dict = df.replace(np.nan, None).to_dict(orient='records')


# Conversione del dizionario in formato JSON
json_data = json.dumps(data_dict, default=str, skipkeys=True)
```

```python
# Salvataggio del JSON in un file
with open('api.json', 'w') as json_file:
    json_file.write(json_data)
```

```python
# Rimozione degli attributi con valore "null"
def remove_null_values(data):
    if isinstance(data, dict):
        return {k: remove_null_values(v) for k, v in data.items() if v is not None}
    elif isinstance(data, list):
        return [remove_null_values(v) for v in data if v is not None]
    else:
        return data

# Rimozione degli attributi nulli dal JSON
json_modified = json.dumps(remove_null_values(json_data), default=str, skipkeys=True)

# Parsing del JSON modificato in un oggetto Python
data = json.loads(json_modified)

with open('api_2.json', 'w') as json_file:
    json_file.write(data)
```

```python
#Carica il file JSON
with open('api_2.json', 'r') as f:
    json_data = json.load(f)


# Itera su ciascun record
for record in json_data:
    # Crea una copia del record
    record_cleaned = record.copy()

    # Rimuovi gli attributi con valore null
    for key, value in record.items():
        if value is None:
            del record_cleaned[key]

    # Sovrascrivi il record originale con il record pulito
    record.clear()
    record.update(record_cleaned)

# Salva il file JSON pulito
with open('api_cleaned.json', 'w') as f:
    json.dump(json_data, f, indent=4)
```
