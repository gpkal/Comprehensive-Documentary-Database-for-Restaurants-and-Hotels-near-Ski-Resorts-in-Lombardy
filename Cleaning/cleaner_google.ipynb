```python
import csv
import re
import pandas as pd
```

```python
# Funzione personalizzata per la sostituzione dei valori
def replace_price(value):
    if '€' in str(value):
        return value
    else:
        return 'price not available'
```

```python
risultati = pd.DataFrame()
```

```python
locations = ['collio', 'rogno', 'valdisotto', 'oltre_il_colle', 'angolo_terme',
           'san_pellegrino_terme', 'valtorta', 'artogne', 'gromo', 'teglio',
           'piuro', 'gerola_alta', 'santa_margherita_di_staffora', 'madesimo',
           'margno', 'valbondione', 'bagolino', 'breno', 'pisogne', 'borno',
           'barzio', 'lecco', 'corteno_golgi', 'moggio', 'schilpario',
           'campodolcino', 'lanzada', 'temu', 'valfurva', 'caspoggio', 'foppolo',
           'castione_della_presolana', 'valleve', 'selvino', 'livigno',
           'ponte_di_legno', 'songavazzo', 'chiesa_in_valmalenco', 'carona',
           'colere', 'valdidentro', 'piazzatorre', 'bormio','aprica']
```

```python
for luoghi in locations:
    df = pd.read_csv(f"rist_{luoghi}.csv")
    df = df.dropna(subset=['OSrXXb'])
    subset = df[['OSrXXb', 'rllt__details', 'yi40Hd', 'RDApEe', 'rllt__details 2', 'rllt__details 3' ]]
    #rinomino le colonne con i nomi che mi servono
    subset = subset.rename(columns = {'OSrXXb':'titles',
                                 'rllt__details': 'type',
                                 'yi40Hd': 'google_ratings',
                                 'RDApEe': 'google_reviews',
                                 'rllt__details 2': 'google_price',
                                  'rllt__details 3': 'locations'})
    subset['città'] = luoghi
    # Sostituire le celle senza il carattere "€" nella colonna "Price"
    subset['google_price'] = subset['google_price'].apply(replace_price)
    # Sostituire le celle NaN con "not available"
    subset = subset.fillna('not available')

    #aggiungo i risulatati in un dataframe
    risultati = pd.concat([risultati, subset], ignore_index = True)
```

```python
risultati
```

```python
# Rimuovi duplicati basati su 'titles', 'type' e 'price'
risultati_senza_duplicati = risultati.drop_duplicates(subset=['titles', 'type', 'google_price'], keep='first')
```

```python
risultati_senza_duplicati
```

```python
risultati_senza_duplicati.to_csv('ristoranti_google.csv', index=False)
```
