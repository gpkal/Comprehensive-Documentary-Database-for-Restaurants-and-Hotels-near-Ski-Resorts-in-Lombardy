## Accuracy of rating column

```python
# Google

# Definisci il pattern regex per il formato del rating
pattern = r'^[0-5],[0-9]$'

# Contatore per il numero di ratings corretti
correct_ratings = 0
wrong_ratings = []

# Itera attraverso i valori della colonna 'rating'
for rating in df1['google_ratings']:
    # Controlla se il rating segue il formato desiderato
    if re.match(pattern, rating):
        correct_ratings += 1
    else:
        wrong_ratings.append(rating)

# Calcola la percentuale di ratings corretti rispetto al totale
accuracy = correct_ratings / len(df1['google_ratings']) * 100
print("Accuracy of ratings:", accuracy)
print(wrong_ratings)
```

```python
# The Fork

# Definisci il pattern regex per il formato del rating
pattern = r'^[1-5]\.\d/5$'

# Contatore per il numero di ratings corretti
correct_ratings = 0

# Itera attraverso i valori della colonna 'rating'
for rating in df2['the_fork_ratings']:
    # Controlla se il rating segue il formato desiderato
    if re.match(pattern, rating):
        correct_ratings += 1

# Calcola la percentuale di ratings corretti rispetto al totale
accuracy = correct_ratings / len(df2['the_fork_ratings']) * 100
print("Accuracy of ratings:", accuracy)
```

```python
# TripAdvisor

# Definisci il pattern regex per il formato del rating
pattern = r'^[1-5]\.\d$'

# Contatore per il numero di ratings corretti
correct_ratings = 0
wrong_ratings = []

# Trasforma i valori nella colonna 'rating' in stringhe
df3['rating'] = df3['rating'].astype(str)

# Itera attraverso i valori della colonna 'rating'
for rating in df3['rating']:

    if isinstance(rating, str):
        # Controlla se il rating segue il formato desiderato
        if re.match(pattern, rating):
            correct_ratings += 1
        else:
            wrong_ratings.append(rating)

# Calcola la percentuale di ratings corretti rispetto al totale
accuracy = correct_ratings / len(df3['rating']) * 100
print("Accuracy of ratings:", accuracy)
print(wrong_ratings)
```

```python
# Booking

# Definisci il pattern regex per il formato del rating
pattern = r'^[1-5]\.\d$'

# Contatore per il numero di ratings corretti
correct_ratings = 0
wrong_ratings = []

# Trasforma i valori nella colonna 'rating' in stringhe
df6['rating'] = df3['rating'].astype(str)

# Itera attraverso i valori della colonna 'rating'
for rating in df6['rating']:
    # Controlla se il rating segue il formato desiderato
    if re.match(pattern, rating):
        correct_ratings += 1
    else:
        wrong_ratings.append(rating)

# Calcola la percentuale di ratings corretti rispetto al totale
accuracy = correct_ratings / len(df6['rating']) * 100
print("Accuracy of ratings:", accuracy)
print(wrong_ratings)
```

## Accuracy of address column

```python
# Ristoranti

# Pattern
pattern_via = r"^(?:Località|Localitá|Rifugio|Lungolago|Corso|Via|Piazza|Frazione|Loc\.?|Viale|Vico|Localita'|Localita|Viale|Piazzale|Contrada|Vicolo|ss) [\w\s\.']+?\d+(?:[a-zA-Z])?$"
pattern_via_interno = r"^(?:Lungolago|Corso|Via|Piazza|Frazione|Loc\.?|Viale|Vico|Localita'|Localita|Viale|Piazzale|Contrada|Vicolo|ss) [\w\s']+?\d+[a-zA-Z]?$"
pattern_via_interno_bis = r"^(?:Lungolago|Corso|Via|Piazza|Frazione|Loc\.?|Viale|Vico|Localita'|Localita|Viale|Piazzale|Contrada|Vicolo|ss) [\w\s]+?\d+[a-zA-Z]?$"
pattern_via_bis = r"^(?:Lungolago|Corso|Via|Piazza|Frazione|Loc\.?|Viale|Vico|Localita'|Localita|Viale|Piazzale|Contrada|Vicolo|ss) [\w\s]+?\d+[a-zA-Z]?\/?\d*[a-zA-Z]?$"
pattern_localita_alt = r'^localita [\w\s]+?\d$'
pattern_apostrofo = r"^(?:Lungolago|Corso|Via|Piazza|Frazione|Loc\.?|Viale|Vico|Localita'|Localita|Viale|Piazzale|Contrada|Vicolo|ss) [\w\s']+?\d$"

# Conter
count_matched = 0
wrong_address = []

# to string
df4['locations'] = df4['locations'].astype(str)

for locations in df4['locations']:
    if re.match(pattern_via, locations, re.IGNORECASE) or re.match(pattern_via_interno, locations, re.IGNORECASE) or \
        re.match(pattern_via_interno_bis, locations, re.IGNORECASE) or re.match(pattern_localita_alt, locations, re.IGNORECASE) or \
        re.match(pattern_apostrofo, locations, re.IGNORECASE) or re.match(pattern_via, locations, re.IGNORECASE):
        count_matched += 1
    else:
        wrong_address.append(locations)


percentage_matched = count_matched / len(df4['locations']) * 100
print("Percentage of matched addresses:", percentage_matched)
print(wrong_address)
```

```python
# Hotel

# Pattern
pattern_via = r"via\s[a-z\.']+[\s\w\.]+\d[\w\/'\-]+"
pattern_viale = r'^viale [\w\s]+?\d$'
pattern_viale_interno = r"^viale [\w\s']+?\d+[a-zA-Z]?$"
pattern_localita = r'^località [\w\s]+?\d$'
pattern_localita_alt = r'^localita [\w\s]+?\d$'
pattern_piazza = r'^piazza [\w\s]+?\d$'
pattern_piazza_apostrofo = r"^piazza [\w\s']+?\d$"
pattern_piazzale = r'^piazzale [\w\s]+?\d$'
pattern_corso = r'^corso [\w\s]+?\d$'
pattern_ss = r'^ss[\w\s]+?\d$'
pattern_via_interno = r'^via [\w\s]+?\d+\/?\w*$'
pattern_via_interno_bis = r'^via [\w\s]+?\d+[a-zA-Z]?$'
pattern_contrada = r'^contrada [\w\s]+?\d$'
pattern_apostrofo = r"^via [\w\s']+?\d$"
pattern_vicolo = r'^vicolo [\w\s]+?\d$'

# Counter
count_matched = 0
matched_address = []
wrong_address = []

df7['Indirizzo'] = df7['Indirizzo'].astype(str)

# Itera attraverso i valori della colonna 'address'
for indirizzo in df7['Indirizzo']:
    # Controlla se l'indirizzo corrisponde a uno dei pattern desiderati (case-insensitive)
    if re.match(pattern_via, indirizzo, re.IGNORECASE) or re.match(pattern_viale, indirizzo, re.IGNORECASE) or \
        re.match(pattern_localita, indirizzo, re.IGNORECASE) or re.match(pattern_localita_alt, indirizzo, re.IGNORECASE) or \
        re.match(pattern_piazza, indirizzo, re.IGNORECASE) or re.match(pattern_corso, indirizzo, re.IGNORECASE) or \
        re.match(pattern_ss, indirizzo, re.IGNORECASE) or re.match(pattern_via_interno, indirizzo, re.IGNORECASE) or \
        re.match(pattern_via_interno_bis, indirizzo, re.IGNORECASE) or re.match(pattern_contrada, indirizzo, re.IGNORECASE) or \
        re.match(pattern_apostrofo, indirizzo, re.IGNORECASE) or re.match(pattern_viale_interno, indirizzo, re.IGNORECASE) or \
        re.match(pattern_piazzale, indirizzo, re.IGNORECASE) or re.match(pattern_piazza_apostrofo, indirizzo, re.IGNORECASE) or \
        re.match(pattern_vicolo, indirizzo, re.IGNORECASE):
        count_matched += 1
        matched_address.append(indirizzo)
    else:
        wrong_address.append(indirizzo)
```

```python
count_non_null = sum(1 for value in matched_address if pd.notnull(value) and value != np.nan)

print("Numero di valori non nulli:", count_non_null)
```

```python
# Calcola la percentuale di vie corrispondenti rispetto al totale
percentage_matched = count_non_null / len(df7['Indirizzo']) * 100
print("Percentage of matched addresses:", percentage_matched)
```
